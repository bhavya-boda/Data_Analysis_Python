## (1) Data Inspection

```python
import pandas as pd

# Load the dataset into a pandas DataFrame
df = pd.read_csv("IMdB_India_Top250.csv")

# a. Display the first 10 rows of the dataframe
df.head(10)
```

```python
# b. Use the info() function to get an overview of the dataset
print("\nDataset Info:")
df.info()
```

```python
# c. Display the summary statistics for numerical columns using describe()
print("\nSummary Statistics:")
print(df.describe())
```

## (2) Data Cleaning

```python
# Identify and handle any missing values in the dataset
print("\nMissing Values:")
print(df.isnull().sum())

# Drop rows or fill missing values as appropriate.
df.dropna(inplace=True)
```

This indicates that we do not have any of the missing values in our dataset.

```python
# Remove any duplicate rows from the dataset
df.drop_duplicates(inplace=True)

# Ensure that the columns related to rankings, years, or ratings are in the correct numerical format.
df['Year of release'] = pd.to_numeric(df['Year of release'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['User reviews'] = pd.to_numeric(df['User reviews'], errors='coerce')

df.info()
```

As we can see  that initially when we had 250 values and now after deleting also we have 250 so there are no duplicate rows.

## (3) Filtering and Sorting

```python
# Filter the dataset to display only movies released after the year 2000
filtered_df = df[df['Year of release'] > 2000]
filtered_df.head()
```

```python
# Sort the DataFrame by IMDB Rating in descending order
sorted_df = df.sort_values(by='Rating', ascending=False)
sorted_df.head()
```

```python
# Create a new DataFrame that contains only movies with an IMDB rating greater than 8.5
high_rated_movies = df[df['Rating'] > 8.5]
high_rated_movies.head()
```

## (4) Group By Operations

```python
# Group the movies by the Year column and calculate the average IMDB rating for each year
average_ratings_by_year = df.groupby('Year of release')['Rating'].mean()
average_ratings_by_year.head()
```

```python
# Group the data by Director and count the number of movies each director has in the top 250
director_movie_count = df.groupby('Director')['Movie name'].count()
director_movie_count.head()
```

```python
# Find the highest-rated movie for each year by grouping the data by Year and selecting the movie with the highest rating in each group
idx = df.groupby('Year of release')['Rating'].idxmax()
highest_rated_movies = df.loc[idx, ['Year of release', 'Movie name', 'Rating']]
highest_rated_movies.head()
```

## (5) Column Creation and Manipulation

```python
# Create a new column named Rating Category that categorizes movies as "Excellent" if the IMDB rating is 9.0 or above, "Good" if between 8.0 and 9.0, and "Average" if below 8.0
df['Rating Category'] = pd.cut(df['Rating'], bins=[0, 8.0, 9.0, float('inf')], labels=['Average', 'Good', 'Excellent'])

# to show variety ans answer used tail instead of head, because in head all movies were almost good
df.tail()
```

```python
# Extract the Year from the movie title (assuming the year is part of the title) and create a new column Extracted Year
df['Extracted Year'] = df['Movie name'].str.extract(r'\((.*?)\)')
print(df.head())

df['Extracted Year'] = df['Movie name'].str.extract(r'\((.*?)\)')
print(df.tail())
```
